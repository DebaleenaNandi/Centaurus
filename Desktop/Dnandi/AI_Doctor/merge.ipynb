```python
import tkinter as tk
import subprocess
import sys
from io import StringIO
import os

# Function to run a Jupyter notebook and capture the output
def run_notebook(notebook_path):
    # Using subprocess to execute Jupyter notebook and capture output
    command = f"jupyter nbconvert --to notebook --execute --inplace {notebook_path}"
    
    # Execute the command
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = process.communicate()

    # If the notebook ran successfully, capture the output
    if process.returncode == 0:
        # Read the output from the notebook's stdout
        return stdout.decode()
    else:
        return stderr.decode()

# Function to create a Tkinter window and display the notebook's output
def create_output_window(notebook_path, title="Notebook Output"):
    output = run_notebook(notebook_path)  # Run the notebook and capture output

    # Create a new Tkinter window
    window = tk.Toplevel()
    window.title(title)

    # Create a Text widget to display the notebook's output
    text_widget = tk.Text(window, height=20, width=80)
    text_widget.pack(padx=10, pady=10)
    text_widget.insert(tk.END, output)  # Insert the output of the notebook

    # Make the text widget read-only
    text_widget.config(state=tk.DISABLED)

# Main function to run the notebooks and open output in separate windows
def run():
    # Create the main window
    root = tk.Tk()
    root.withdraw()  # Hide the root window since we don't need it

    # Define the paths to your Jupyter notebooks
    notebook1_path = "C:/Users/Dnandi/Desktop/Dnandi/AI_Doctor/AI_Doctor.ipynb"  # Path to the first notebook
    notebook2_path = "C:/Users/Dnandi/Desktop/Dnandi/AI_Doctor/Lab_Test.ipynb"  # Path to the second notebook
   

    # Run the notebooks and display the output in separate windows
    create_output_window(notebook1_path, "Notebook 1 Output")
    create_output_window(notebook2_path, "Notebook 2 Output")

    # Start the Tkinter main loop to keep the windows open
    root.mainloop()

# Run the Tkinter GUI
run()
```
